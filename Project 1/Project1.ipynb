# 1. Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# 2. Generate Dataset (150 records)

```python
np.random.seed(42)

n = 150

data = pd.DataFrame({
    "Household_ID": [f"HH{i+1:03d}" for i in range(n)],
    "Age_of_Household_Head": np.random.randint(25, 70, n),
    "Household_Income": np.random.randint(20000, 150000, n),
    "Education_Level": np.random.choice(["Primary", "Secondary", "Graduate", "Post-Graduate"], n),
    "Family_Size": np.random.randint(1, 8, n),
    "Owns_House": np.random.choice(["Yes", "No"], n),
    "Urban_Rural": np.random.choice(["Urban", "Rural"], n)
})

# data.head()
```

# 4. Central Tendency

```python
# Income
income_mean = data["Household_Income"].mean()
income_median = data["Household_Income"].median()
income_mode = data["Household_Income"].mode()[0]

print("Income Mean:", income_mean)
print("Income Median:", income_median)
print("Income Mode:", income_mode)
```

```python
# Age
age_mean = data["Age_of_Household_Head"].mean()
age_median = data["Age_of_Household_Head"].median()
age_mode = data["Age_of_Household_Head"].mode()[0]

print("Age Mean:", age_mean)
print("Age Median:", age_median)
print("Age Mode:", age_mode)
```

```python
# income summery

if income_mean > income_median:
    print("Income distribution is positively skewed (right skewed).")
elif income_mean < income_median:
    print("Income distribution is negatively skewed (left skewed).")
else:
    print("Income distribution is approximately symmetric.")
  
```

# 5. Measures of Dispersion

```python
# Range
income_range = data["Household_Income"].max() - data["Household_Income"].min()

# Variance
income_variance = data["Household_Income"].var()

# Standard Deviation
income_std = data["Household_Income"].std()

print("Range:", income_range)
print("Variance:", income_variance)
print("Std Dev:", income_std)
```

```python
# IQR
Q1 = data["Household_Income"].quantile(0.25)
Q3 = data["Household_Income"].quantile(0.75)
IQR = Q3 - Q1

print("Q1:" , Q1)
print("Q3:" , Q3)
print("IQR:", IQR)
```

# 6. Distribution

```python
# Histogram + KDE

plt.figure()
sns.histplot(data["Household_Income"], kde=True)
plt.title("Income Distribution")
plt.xlabel("Income")
plt.ylabel("Frequency")
plt.show()
```

```python
# Gaussian Curve 

plt.figure()
sns.regplot(x="Age_of_Household_Head", y="Household_Income", data=data)

plt.title("Age vs Income Relationship")
plt.show()
```

```python
#  Skewness & Kurtosis

skewness = data["Household_Income"].skew()
kurtosis = data["Household_Income"].kurt()

print("Skewness:", skewness)
print("Kurtosis:", kurtosis)
```

# 7. Data Categorization


```python
# Boxplot (Members vs Education)

plt.figure()
sns.boxplot(x="Education_Level", y="Family_Size", data=data)
plt.title("Members by Education Level")
plt.xticks(rotation=45)
plt.show()
```

```python
# Scatter Plot (Age vs Income)

plt.figure()
sns.scatterplot(x="Age_of_Household_Head", y="Household_Income", data=data)
plt.title("Age vs Income")
plt.show()
```
