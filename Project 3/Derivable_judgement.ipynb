# Part A



### 1. What is Inferential Statistics?

Inferential statistics is the branch of statistics used to make predictions or conclusions about a population based on sample data.
It helps in decision-making using techniques like hypothesis testing and confidence intervals.

### 2. What is Hypothesis Testing and its Components?

Hypothesis testing is a statistical method used to check whether a claim about a population is true or not using sample data.
Its main components are Null Hypothesis (H₀), Alternative Hypothesis (H₁), significance level (α), test statistic, and p-value.

### 3. Explain Confidence Interval and Critical Value.

A confidence interval is a range of values within which the true population parameter is expected to lie with a certain confidence level (like 95%).
A critical value is the cutoff point used to decide whether to reject the null hypothesis in a statistical test.

### 4. Define p-value.

The p-value is the probability of getting the observed result, or more extreme results, if the null hypothesis is true.
A small p-value (usually less than 0.05) indicates strong evidence against the null hypothesis.

### 5. Differentiate Type I and Type II Errors.

Type I Error occurs when a true null hypothesis is rejected incorrectly (false positive).
Type II Error occurs when a false null hypothesis is accepted incorrectly (false negative).

### 6. Brief Description of z-test, t-test, chi-square test, and ANOVA test.
- z-test: Used to compare means when sample size is large and population standard deviation is known.
- t-test: Used to compare means when sample size is small and population standard deviation is unknown.
- Chi-square test: Used to test relationships between categorical variables.
- ANOVA test: Used to compare means of three or more groups.

### 7. What is Covariance?

Covariance measures the relationship between two variables and shows whether they move together or in opposite directions.
A positive covariance means variables increase together, while a negative covariance means one increases when the other decreases.

### 8. What is Correlation?

Correlation is a statistical measure that shows the strength and direction of the relationship between two variables.
It tells us whether variables increase or decrease together.

- Positive Correlation: Both variables move in the same direction.
- Negative Correlation: Variables move in opposite directions.
- Zero Correlation: No relationship between variables.

# 2️. Confidence Interval Calculation

```python
# Age Data
age_mean = df["age"].mean()
age_std = df["age"].std()
n = len(df)

# 95% Confidence Interval
confidence = 0.95

age_ci = stats.norm.interval(
    confidence,
    loc=age_mean,
    scale=age_std/np.sqrt(n)
)


print("Mean Age:", age_mean)
print("95% Confidence Interval for Age:", age_ci)
```

```python
# Weight Data

weight_mean = df["weight"].mean()
weight_std = df["weight"].std()

weight_ci = stats.norm.interval(
    confidence,
    loc=weight_mean,
    scale=weight_std/np.sqrt(n)
)

print("Mean Weight:", weight_mean)
print("95% Confidence Interval for Weight:", weight_ci)
```

# 3. Critical Value and p-value

```python
from statsmodels.stats.weightstats import ztest

z_stat, p_value = ztest(df["bmi"], value=25)

print("Z-Statistic:", z_stat)
print("P-Value:", p_value)
```

```python
# Interpretation

if p_value < 0.05:
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")

```

# 4. Z-Test or T-Test

```python
from scipy.stats import norm

# Z test

# Groups
smoker = df[df["smoking_status"]=="Smoker"]["glucose_level"]
non_smoker = df[df["smoking_status"]=="Non-Smoker"]["glucose_level"]

# Statistics
X1 = smoker.mean()
X2 = non_smoker.mean()

sigma1 = smoker.std()
sigma2 = non_smoker.std()

n1 = len(smoker)
n2 = len(non_smoker)

# Z-score
z = (X1 - X2) / np.sqrt((sigma1**2)/n1 + (sigma2**2)/n2)

# P-value
p = 2 * (1 - norm.cdf(abs(z)))

# Critical value
critical_value = 1.96

print("Z-Score:", round(z,3))
print("Critical Value:", critical_value)
print("P-Value:", round(p,4))
```

```python
# Interpretation

if abs(z) > critical_value:
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
```

```python
# T test

from scipy.stats import ttest_ind

male_bmi = df[df["gender"]=="Male"]["bmi"]
female_bmi = df[df["gender"]=="Female"]["bmi"]

t_stat, p_value = ttest_ind(male_bmi, female_bmi)

print("T-Statistic:", t_stat)
print("P-Value:", p_value)
```

```python
# Interpretation

alpha = 0.05

if p_value < alpha:
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
```

# 5. Chi-Square Test

```python
from scipy.stats import chi2_contingency

table = pd.crosstab(df["smoking_status"], df["diabetes"])

chi2, p, dof, expected = chi2_contingency(table)

print("Chi-Square Value:", chi2)
print("P-Value:", p)
```

```python
# Interpretation

if p < 0.05:
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
```

# 6. ANOVA Test

```python
from scipy.stats import f_oneway

g1 = df[df["age_group"]=="18-25"]["glucose_level"]
g2 = df[df["age_group"]=="26-35"]["glucose_level"]
g3 = df[df["age_group"]=="36-45"]["glucose_level"]
g4 = df[df["age_group"]=="46-60"]["glucose_level"]

f_stat, p_value = f_oneway(g1, g2, g3, g4)

print("F-Statistic:", f_stat)
print("P-Value:", p_value)
```

```python
#Interpretation

if p_value < 0.05:
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
```
