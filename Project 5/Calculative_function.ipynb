### **2. Norms , Dot product and cross product**

-  **Norms -**
Norms are used to measure the magnitude or length of a vector.

- **Dot Product-**
The dot product measures the similarity between two vectors.

- **Cross Product -**
The cross product produces a vector that is perpendicular to two three-dimensional vectors.

```python

# Norms
norm1 = np.linalg.norm(s1, 1)
norm2 = np.linalg.norm(s1, 2)

# Dot product
dot_product = np.dot(s1, s2)

# Cross product
cross_product = np.cross(s1, s2)

print("norm1 :" , norm1 )
print("norm2 :" , norm2 )
print("dot_product :" , dot_product )
print("cross_product :" , cross_product )
```

### **3. Projection of One Vector onto Another**

- Vector projection shows the component of one vector in the direction of another vector.

```python
projection = (np.dot(s1, s2) / np.dot(s2, s2)) * s2

print("Projection :" ,projection)
```

## Part D: Eigenvalues & Decomposition

### **7. Eigenvalues and Eigenvectors**

Eigenvalues represent the amount of variance captured in a particular direction, while eigenvectors represent the directions of maximum variance in the data.

```python
# Covariance Matrix
cov_matrix = np.cov(matrix2.T)

eigenvalues , eigenvactor = np.linalg.eig(cov_matrix)
print("Eigenvalues : \n" , eigenvalues , "\n" )
print("Eigenvactor : \n" , eigenvactor , "\n" )
```

### **8. LU Decomposition**

LU decomposition factorizes a matrix into a lower triangular matrix and an upper triangular matrix.


```python
from scipy.linalg import lu

P, L, U = lu(matrix2)

print("Permutation Matrix (P):")
print(P)

print("\nLower Triangular Matrix (L):")
print(L)

print("\nUpper Triangular Matrix (U):")
print(U)
```

## Interpretation

LU decomposition simplifies matrix operations by decomposing the original matrix into lower and upper triangular matrices.

### **9. Singular Value Decomposition (SVD)**

SVD decomposes a matrix into three matrices and is widely used for dimensionality reduction and data compression.

```python
U, S, VT = np.linalg.svd(matrix2)

print("U Matrix:")
print(U)

print("\nSingular Values:")
print(S)

print("\nV Transpose Matrix:")
print(VT)
```

## Interpretation

SVD identifies the most important patterns in the dataset and helps reduce dimensionality while preserving essential information.

# Part E: Dimensionality Reduction

### **10. Principal Component Analysis (PCA)**

PCA is a dimensionality reduction technique that transforms data into a smaller set of uncorrelated variables while preserving most of the variance.

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)

pca_result = pca.fit_transform(matrix2)

print("PCA Result:")
print(pca_result)
```

## Interpretation

PCA reduced the dataset from three dimensions to two dimensions while retaining most of the important information.

### **9.Linear Discriminant Analysis (LDA)**

LDA is a supervised learning technique used to separate data into predefined classes.

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

student_means = matrix2.mean(axis=1)

overall_mean = student_means.mean()

labels = np.where(
    student_means >= overall_mean,
    "Above Average",
    "Below Average"
)

lda = LinearDiscriminantAnalysis(n_components=1)

lda_result = lda.fit_transform(matrix2, labels)

print(lda_result)
```
