# 1. Understanding the Basics

### What is Probability?

Probability is a branch of mathematics that measures the likelihood of an event occurring.
The value of probability always lies between 0 and 1.

Probability = 0 → Impossible event
Probability = 1 → Certain event

Formula:

P(E)=
Total Outcomes
Favorable Outcomes
	​

### Key Probability Terminology
- Experiment

An activity that produces outcomes.

Example:
Selecting a student from the dataset.

- Outcome

The result of an experiment.

Example:
The selected student passes the exam.

- Event

A collection of outcomes.

Example:
Students having attendance greater than 80%.

- Sample Space

The set of all possible outcomes.

Example:
All students in the dataset.


### Three Probability Event Examples from Dataset
Event 1

Student studies more than 10 hours/week.

Event 2

Student attendance is greater than 80%.

Event 3

Student passes the final exam.

# 2. Types of Events

# 1.. Empirical Probability

Empirical probability actual dataset observations par based hoti hai.

- Scenario

Find the probability that a student passes the final exam.

Formula:

P(E) = Total outcomes / Number of favorable outcomes
​


```python
# Empirical Probability

pass_students = len(data[data['final_exam_pass'] == 'Pass'])

total_students = len(data)

empirical_probability = pass_students / total_students

print("Students Passed:", pass_students)
print("Total Students:", total_students)
print("Empirical Probability of Passing:", empirical_probability)
```

# 2. Theoretical Probability

Theoretical probability assumptions par based hoti hai.

- Scenario

Assume there are only two equally likely outcomes:

Pass
Fail

Then probability of passing is:

P(Pass)= 2 / 1 = 0.5

```python
# Theoretical Probability

# Possible outcomes
outcomes = ['Pass', 'Fail']

# Favorable outcome
favorable_outcomes = 1   # Pass

# Total outcomes
total_outcomes = len(outcomes)

# Theoretical probability
theoretical_probability = favorable_outcomes / total_outcomes

print("Favorable Outcomes:", favorable_outcomes)
print("Total Outcomes:", total_outcomes)
print("Theoretical Probability of Passing:", theoretical_probability)
```

# 3. Random Variable & Probability Distribution
 

- Random Variable

A random variable is a numerical value assigned to outcomes of a random experiment.

- Example:
Number of students passing among 3 randomly selected students.

```python

# Convert Pass/Fail into numeric values
# Pass = 1
# Fail = 0

data['pass_numeric'] = data['final_exam_pass'].map({
    'Pass': 1,
    'Fail': 0
})

print(data[['final_exam_pass', 'pass_numeric']].head())

# Randomly select 3 students many times
samples = []

for i in range(1000):
    
    # Select 3 random students from My dataset
    sample = data['pass_numeric'].sample(n=3, replace=True)
    
    # Count how many students passed
    passed_count = sample.sum()
    
    # Store result
    samples.append(passed_count)

# Create Probability Distribution Table

distribution = pd.Series(samples).value_counts(normalize=True).sort_index()

distribution_table = pd.DataFrame({
    'X (Students Passed)': distribution.index,
    'P(X)': distribution.values
})

print("\nProbability Distribution Table:\n")
print(distribution_table)

# Mean Calculation

mean = np.mean(samples)

# Variance Calculation

variance = np.var(samples)

print("\nMean:", mean)
print("Variance:", variance)
```

# 4. Venn Diagram in Probability

```python
# Set A
A = set(data[data['study_hours'] > 10].index)

# Set B
B = set(data[data['attendance'] > 80].index)

# Count Values
only_A = len(A - B)

only_B = len(B - A)

both_A_B = len(A & B)

print("Students studying more than 10 hours:", len(A))
print("Students with attendance > 80%:", len(B))
print("Students satisfying both conditions:", both_A_B)

# Draw Venn Diagram

plt.figure(figsize=(6,6))

venn2(
    subsets=(only_A, only_B, both_A_B),
    set_labels=('Study > 10 Hours', 'Attendance > 80%')
)

plt.title("Venn Diagram of Student Conditions")

plt.show()
```

# 5. Contingency Table & Probability Calculations

A contingency table is used to analyze the relationship between two categorical variables.

- In this project:

Variable 1 → group_discussion (Yes/No)
Variable 2 → final_exam_pass (Pass/Fail)

- Using this table, we calculate:

Joint Probability
Marginal Probability
Conditional Probability

-> Formula

- Joint Probability

P(A∩B) = Total Students / Students in both A and B	​

- Marginal Probability

P(A) = Total Students / Students satisfying A
	​


```python

# Create Contingency Table

contingency_table = pd.crosstab(
    data['group_discussion'],
    data['final_exam_pass']
)

print("Contingency Table:\n")
print(contingency_table)

# Total Students
total_students = len(data)

# -----------------------------------
# Joint Probability
# Participates in discussion AND Pass
# -----------------------------------

joint_count = len(
    data[
        (data['group_discussion'] == 'Yes') &
        (data['final_exam_pass'] == 'Pass')
    ]
)

joint_probability = joint_count / total_students

print("\nJoint Probability:")
print(joint_probability)

# -----------------------------------
# Marginal Probability
# Probability of Passing Exam
# -----------------------------------

pass_count = len(
    data[data['final_exam_pass'] == 'Pass']
)

marginal_probability = pass_count / total_students

print("\nMarginal Probability:")
print(marginal_probability)

# -----------------------------------
# Conditional Probability
# Pass given Discussion = Yes
# -----------------------------------

discussion_yes = len(
    data[data['group_discussion'] == 'Yes']
)

conditional_probability = joint_count / discussion_yes

print("\nConditional Probability:")
print(conditional_probability)
```

# 6. Understanding Relationships

Conditional probability helps us understand how one event affects another event.

-> In this project:

Event A → Student passes the exam
Event B → Student participates in group discussion

### Interpretation in Plain Language

The conditional probability result shows that students participating in group discussions are more likely to pass the final exam.

This means participation in group discussions positively affects exam performance.

### Independent or Dependent Events ?
->
Answer: Dependent Events

- Reason

If two events are independent, then one event does not affect the probability of another event.

But in this dataset:

Students participating in discussions have a higher probability of passing.
Therefore, passing the exam depends on participation in group discussions.

Hence, both events are dependent events.

### Mutually Exclusive?
Answer: No

The events are not mutually exclusive because:

A student can both participate in group discussions AND pass the exam at the same time.

Therefore:

They can occur together.
So they are not mutually exclusive events.
